## What exactly is a function?

You've seen functions and used them. They include ``print()``, ``forward()`` and ``range()``, and they provide useful _functionality_.

But what exactly is a function?

In programing, a function allows your to wrap up a bunch of code into a a single call. You can pass a function _parameters_ and it can pass back a _return value_ when it is done. The best thing about a function is that whoever calls it doesn't need to worry about _how_ it works... they can trust that the function _will work_.

Making a function is actually really easy. Let's have a look...

````
# Here is a function that takes a name and a language code, and returns an 
# appropriate greeting.

def universal_greeting(name, language) :
  if language in ('EN', 'en') :
    return "Hello " + name + "!"
  elif language in ('FR', 'fr') :
    return "Bonjour " + name + '!'
  elif language in ('DE', 'de') :
    return "Hallo " + name + '!'
  elif language in ('PT', 'pt') :
    return 'Olá ' + name + '!'
  elif language in ('ES', 'es') :
    return '¡Hola ' + name + '!'
  else :
    return 'Hi ' + name + '.'
````

The function above takes two parameters: ``name`` and ``language``. It then uses an ``if``...``elif``...``else`` statment to determine an appropriate greeting message to return. If the language is not recognized, the function returns "Hi _name_."

Let's look at how to call this simple function:

````
# Run this cell to test the function and cature it's return value in output.
output = universal_greeting('Nugget', 'es')

# Print output to see what the function returned.
print("The function returned: " + output)
````

And _voilà_! We wrote a function.

## Triangle function

We've seen how functions can take parameters and work with that information to do something simple. Now let's develop a function together that will draw a triangle. We'll use what we learnt earlier in this discovery about drawing triangles, and we will see how functions can help us make lots of different triangles.

Let's start by defining a simple triangle drawing function in the cell below.

````
# Define a simple function to draw a triangle
def draw_triangle() :
  # Use a loop to draw each side and turn at each corner
  for count in range(3) :
    turtle.forward(30)
    turtle.right(120)
````

````
# Now test the draw_triangle() function by calling it.
draw_triangle()
````

It worked! The function drew a simple triangle for us.

But the triangle it drew isn't very interesting, is it? It's small and white... wouldn't it be nice if our function could draw a bigger (or smaller) triangle, in a colour that we choose, and maybe have it point in a direction of our choice?

We can accomplish all this by adding parameters to our ``draw_triangle()`` function. Let's do that below.

````
# A function to draw a triangle given the size, colour and how much to turn 
# before starting.
def draw_triangle(side_length, colour, turn) :
  # Set the colour
  turtle.color(colour)
  
  # Turn the turtle right
  turtle.right(turn)

  # Use a loop to draw each side and turn at each corner
  for count in range(3) :
    turtle.forward(side_length)   # Move forward by the given length
    turtle.right(120)
````

````
# Let's test the new draw_triangle() function by calling it a few times.
draw_triangle(60, 'green', 30)

draw_triangle(105, 'blue', 100)

draw_triangle(38, 'yellow', 200)
````

Now that is a much cooler function for drawing triangles! Look at all the varieties of triangles we can make! Hey, I bet we could even use the function to make cool patterns &ndash; try running this:

````
# Clear the canvas
turtle.initializeTurtle()

# Make a pattern of triangles!
for count in range(36) :
  draw_triangle(count*10, 
                turtle.VALID_COLORS[count % len(turtle.VALID_COLORS)],
                count)
````

I really like the fun things we can do with our new ``draw_triangle()`` function, but I miss how the old function was so simple to call. Remember how you could just call ``draw_triangle()`` and turtle would draw a simple, white triangle without having to worry about the telling it a size, colour or turn?

Wouldn't it be great if our triangle function could do all the cool stuff if we passed it parameters, but could also just draw a nice simple triangle if we didn't pass any parameters?

Well, it can &ndash; if we define the function with _default parameter values_. It works like this:

````
# A triangle function twith default parameter values.
def draw_triangle(side_length=30, colour='yellow', turn=30) :
  # Set the colour
  turtle.color(colour)
  
  # Turn the turtle right
  turtle.right(turn)

  # Use a loop to draw each side and turn at each corner
  for count in range(3) :
    turtle.forward(side_length)   # Move forward by the given length
    turtle.right(120)
````

````
# Clear the canvas
turtle.initializeTurtle()

# Test calling the function without parameters
draw_triangle()
````

Nice and simple! Now let's try passing parameters to the function. But I don't want to pass all the parameters, just colour and turn. To help the function know what values I mean to pass, I will provide them as _named parameters_.

````
# Call the function passing named parameters for colour and turn
draw_triangle(turn=75, colour='blue')
````
