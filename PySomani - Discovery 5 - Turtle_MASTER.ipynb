You can instruct the turtle to move, draw, turn and change colour. Run the cells below to learn how.

````
# Clear the canvas and reset the turtle to the middle.
turtle.initializeTurtle()
````

Turtle can move using the ``forward()`` and ``backward()`` functions. You specify the distance in units, which the turtle will translate into screen pixels.

Run the cell below to move the turtle forward by 100 units.

````
# Move forward by 100 units
turtle.forward(100)
````

To change direction, you tell the turtle to swivel right or left. But how far to the right or left should she turn? You tell her that by specifying the number of _degrees_.

>**What are degrees?**
>
>Don't worry if you haven't yet learnt about angles and degrees. When telling the turtle to turn, degrees are just a way to tell it how much to turn.
> 
>There are 360&deg; in a full turn. That means that if you tell turtle to turn right or left by 360&deg;, it will end up facing the same direction as it was before turning. So if you want her to face backwards, you should turn her half that amount &ndash; that's 180&deg; (360&deg; divided by 2). A quarter turn is 90&deg; (360&deg; divided by 4). You can use the circle diagram below to see how many degrees to specify base on how much you want turtle to turn.
>
><a title="Autiwa / CC BY-SA (https://creativecommons.org/licenses/by-sa/3.0)" href="https://commons.wikimedia.org/wiki/File:Rapporteur.svg"><img width="256" alt="There are 360 degrees in a full rotation." src="https://upload.wikimedia.org/wikipedia/commons/thumb/3/3f/Rapporteur.svg/256px-Rapporteur.svg.png"></a>

````
# Let's turn turtle to the left by 90 degrees...
turtle.left(90)

# ...and move forward by 50 units
turtle.forward(50)
````

## Shapes and loops

Have a look at the code that we've written for square and triangle. Could have used our coding knowledge to accomplish the same thing with fewer lines of code?

You probably noticed that we defined a few variables that we didn't end up using, and that there is a lot of repeted code. Ignore the extra variables for the moment &ndash; let's see what we can do to get rid of that repeated code.

In our last discovery, we learnt how to use for-loops and while-loops to make Python repeat our instructions. We can use a loop here to make coding the square less repetitious.

````
# These are the variables we defined for the square.
square_sides = 4  # Number of sides in a square
side_length = 50  # Set this to however long we want each side to be
corners = 4       # Number of corners in a square
square_turn = 90  # Number of degrees that the turtle needs to turn at each corner
side_colour = 'yellow'  # The colour we've chosen for our square.

# Start by clearning the canvas
turtle.initializeTurtle()

# Set the colour
turtle.color(side_colour)
````

So far, so good. Looking back at our square code above, what part of it was repeated over and over again? 

Of course, it was drawing the side and turning the corner. Let's use a for-loop to accomplish the repetition.

````
# We need to repeat the loop enough times to draw each side of our square
for count in range(4) :
  # Draw a side
  turtle.forward(side_length)

  # Turn left at the corner
  turtle.left(square_turn)
````

Look at that &ndash; we just reduced eight lines of code to three by using a for-loop. See if you can do the same below for your triangle code.
