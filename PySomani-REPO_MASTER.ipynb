# Update the pySomani-master repo

## Set up and pull the latest

```
%%bash
rootdir="/content/gdrive/My Drive"
psdir="$rootdir/PySomani_MASTER"

reponame="pySomani-master"
giturl="https://github.com/ayazs/$reponame"
gitdir="$psdir/$reponame"

[ ! -d "$psdir" ] && mkdir -p "$psdir"

cd "$psdir"

git config --global user.email "ayaz.somani@gmail.com"
git config --global user.name "Ayaz Somani"

[ ! -d "$gitdir" ] && git clone "$giturl"

cd "$gitdir"

# Set the origin push / pull URLs
git remote set-url origin $giturl

# Pull the latest from the repo
git pull

# Show git status
git status
```

## Add files to the repo

```
%%bash
rootdir="/content/gdrive/My Drive"
psdir="$rootdir/PySomani_MASTER"

reponame="pySomani-master"
giturl="https://github.com/ayazs/$reponame"
gitdir="$psdir/$reponame"

git config --global user.email "ayaz.somani@gmail.com"
git config --global user.name "Ayaz Somani"

cd "$gitdir"

# Set the origin push / pull URLs
git remote set-url origin $giturl

# Add new folders and files here
git add "PySomani-SETUP-UPDATE_MASTER.ipynb"

# Show git status
git status
```

## Commit changes

```
%%bash
rootdir="/content/gdrive/My Drive"
psdir="$rootdir/PySomani_MASTER"

reponame="pySomani-master"
giturl="https://github.com/ayazs/$reponame"
gitdir="$psdir/$reponame"

git config --global user.email "ayaz.somani@gmail.com"
git config --global user.name "Ayaz Somani"

cd "$gitdir"

# Set the origin push / pull URLs
git remote set-url origin $giturl

# Set commit message here
msg="Cleared outputs and updated GDrive preview."

# Call git commit to commit any changes
git commit -m "$msg"

# Display current git status
git status
```

## Push to remote

```
%%bash
rootdir="/content/gdrive/My Drive"
psdir="$rootdir/PySomani_MASTER"

reponame="pySomani-master"
gituser="ayazs"
gittoken="5bc3de321be347773a6e950d937b9bfa68170922"
giturl="https://$gituser:$gittoken@github.com/ayazs/$reponame"
#giturl="https://github.com/ayazs/$reponame"
gitdir="$psdir/$reponame"

git config --global user.email "ayaz.somani@gmail.com"
git config --global user.name "Ayaz Somani"

cd "$gitdir"

# Set the origin push URL
git remote set-url --push origin $giturl
git remote show origin

# This pushes any committed changes to the remote github repo
git push origin master

# Display git status
git status
```
